# determine the earliest time for each event

```python
data.groupby('Event name')['total_hour'].min().sort_values()
```

```python
event_name=[]
par_id=[]
time=[]
age=[]
gender=[]
speed=[]
age_cat=[]
length=[]
club_=[]
gr=data.groupby('Event name')
for event,dat_ in gr:
    
    short_time=dat_['total_hour'].min()
    number=dat_[dat_['total_hour']==short_time].shape[0]
    data_=dat_[dat_['total_hour']==short_time].reset_index()
   
    

    for i in range(0,number):
        id=data_['Athlete ID'][i]
        
           
            
            
        particapant_age=data_[data_['Athlete ID']==id]['Athlete age category'].values[0]
            
        particapant_gender=data_[data_['Athlete ID']==id]['Athlete gender'].values[0]
        particapant_speed=data_[data_['Athlete ID']==id]['Athlete average speed'].values[0]
        particapant_age_cat=data_[data_['Athlete ID']==id]['Athlete age category'].values[0]
        race_length=data_[data_['Athlete ID']==id]['Event distance/length'].values[0]
        club=data_[data_['Athlete ID']==id]['Athlete club'].values[0]
                        
        event_name.append(event)
        par_id.append(id)
        time.append(short_time)
        age.append(particapant_age)
        gender.append(particapant_gender)
        speed.append(particapant_speed)
        age_cat.append(particapant_age_cat)
        length.append(race_length)
        club_.append(club)
            
    
        
  
                                  
           
           
winner_data=pd.DataFrame({'Evenet_name':event_name,'ID':par_id,'Time':time,'Age':age,'Gender':gender,'Speed':speed,'Age_cat':age_cat,'Length':length,'Club':club_})

winner_data       
                                                                                                                    
```
